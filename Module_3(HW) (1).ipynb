#1.Line Chart

```python
import matplotlib.pyplot as plt
#define dataset
info= {
    "Months":["Jan", "Feb", "March", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"],
    "Unit":[10,15,18,16,11,20,25,26,28,30,32,33],
    "Revenue":[1000,1500,1800,1600,1100,2000,2500,2600,2800,3000,3200,3300]
}
#Line Chart
Months = info["Months"]
Revenue = info["Revenue"]
#plot line charts
plt.plot(Months,Revenue,marker="o",linestyle="--",linewidth=2)
#Add label and title in charts
plt.xlabel("Months", fontsize=14, color="blue")
plt.ylabel("Revenue", fontsize=14, color="Green")
plt.title("Months vs Revenue", fontsize=16, color="orange")
#add grids
plt.grid(True)
#show the charts
plt.show()
```

#2.Bar Chart

```python
import matplotlib.pyplot as plt

Months = ["Jan", "Feb", "March", "Apr", "May", "June"]
Unit = [10,15,18,16,11,20]
#Bar chart
plt.bar(Months,Unit,color=["pink","blue","orange","violet","green","red"],edgecolor="black",linestyle="--")
#label
plt.title("Months vs Unit", color="blue")
plt.xlabel("Months", color="red")
plt.ylabel("Unit", color="red")
plt.xticks(rotation=45)
#grids
plt.grid(axis="y")
plt.show()
```

#3.Histogram

```python
import matplotlib.pyplot as plt

Units =[50,56,72,88,98,60,62,74,92,90]
#Add color and Edge color
#use alpha for light and dark shade color
plt.hist(Units, bins=8, edgecolor="red",color="skyblue", alpha=0.9)
plt.title("Units distribution")
plt.xlabel("Units")
plt.ylabel("frequency")
#use grid for y axis
plt.grid(axis="y",color="blue",linestyle="--")
#Add background color
plt.gca().set_facecolor("pink")
plt.show()
```

#4.Pie Chart

```python
#data distribution
Quarters = ["Q1:[1:3]", "Q2:[3:6]", "Q3[6:9]", "Q4[9:12]"]
sum= [4,9,15,21]
#means add any color whatever we want
Rang= ["red", "yellow", "orange", "skyblue"]
#pie chart
plt.pie(sum, labels=Quarters, autopct= "%1.1f%%", startangle=30, explode=(0.2,0.3,0.1,0.5),shadow=True, colors= Rang)
#title
plt.title("sum vs Quarters")
#creating pie chart circle
plt.axis("equal")
#add legend
plt.legend(Quarters, title= "vs sum")
plt.show()
```

#5.Scatter

```python
import matplotlib.pyplot as plt

#Data
Months = ["Jan","Feb","March","April","May"]
Unit = [10,15,18,16,11]
#add size
size=[100,300,600,800,1000]
#add colors
colors=["red","green","pink","orange","grey"]
plt.scatter(Months, Unit, s=size, c= colors, alpha= 0.8)
#add intensity
plt.xlabel("Months")
plt.ylabel("Unit")
plt.title("Months vs Unit")
plt.show()
```
